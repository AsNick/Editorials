## Задача А: Покупка Испаньолы.


Условие и вопрос задачи задачи предельно понятны:нужно определить четность суммы делителей числа $N$.   
**Наивное решение**: найти все делители числа $N$, просуммировать их и посмотреть четность суммы. Однако, на ограничениях задачи ($N \leq 10^{18})$ такое решение получит TL.

**Полное решение.**

Рассмотрим разложение числа $N$ на простые множители:

$N = p_1^{k_1}p_2^{k_2}p_3^{k_3}...p_m^{k_m}$

Имеем два случая:  
+ $p_1 = 2$
+ $p_1 \neq 2$

Рассмотрим второй случай. Имея разложение на простые множители, можно найти все делители числа. В нашем случае все делители будут нечетные. Сколько таких делителей? Делители числа идут парами, кроме случая, когда число - полный квадрат, в этом случае один делитель непарный. Значит, если число, в разложении которого на простые множители отсутствует степент двойки, не является полным квадратом , оно имеет четное количество нечетных делителей, которые в сумме дают четное число. Если же число является полным квадратом, то мы имеем нечетное количество нечетных делителей, которые в сумме дают нечетное число.

Рассмотрим первый случай и сведем его ко второму. Число $N$, в разложении которого на простые множители есть степень двойки, имеет четные делители. Их количество не важно, в сумме они дают четное число. Если $N$ не степень двойки, в его разложении будут и нечетные множители.   
$N = 2^{k_1}p_2^{k_2}p_3^{k_3}...p_m^{k_m}$  
Пусть $N_1 = p_2^{k_2}p_3^{k_3}...p_n^{k_n}$.   
Числа такого вида рассмотрены во втором случае. Если $N_1$ не полный квадрат - сумма его делителей четна, в противном случае - нечетна.

**План решения задачи А:** делим число$N$ на два пока это возможно. Если полученное после этой процедуры число является полным квадратом, то не получится купить Испаньолу, иначе - получится.

```python
n = int(input())
while n % 2 == 0:
    n //= 2

m = int(n ** .5)
if m * m == n:
    print('NO')
else:
    print('YES')
    
```
