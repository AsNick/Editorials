## Задача J: Пепси-Кола в баллонах.

Решим предварительно простую задачу из курса школьной математики:
на числовой прямой  целочисленными координатами своих концов заданы два отрезка $a =\left [ l_1, r_1\right ]$ и $b =\left [l_2, r_2 \right ]$. Нужно найти пересечение этих отрезков.

Решение: пусть $l = min(l_1, l_2), r = max(r_1, r_2)$, тогда:  
+ если $l > r$ то $ {a} \cap {b} = {\varnothing }$  
+ если $l = r$ то ${a} \cap {b} = l$  
+ если $l < r$ то ${a} \cap {b}
= \left [ l, r\right ]$  
  
  
Задача решена.

Теперь внимательно посмотрим на задачу "Пепси-Кола в баллонах" и увидим, что она ничем не отличается от задачи о пересечении отрезков.

Действительно, $N_1$ - количество баллонов ёмкостью $50$ литров,  задает нам отрезок $A = \left [ V_{A_l}, V_{A_r}\right ]$, где $V_{A_l}$ и $V_{A_r}$ - минимальный и максимальный объем пепси-колы, который можно привезти на съемочную площадку в пятидесятилитровых баллонах. 

Количество баллонов ёмкостью $70$ литров $N_2$ задает нам отрезок $B = \left [ V_{B_l}, V_{B_r}\right ]$, где $V_{B_l}$ и $V_{B_r}$ - минимальный и максимальный объем пепси-колы, который можно привезти на съемочную площадку в семидесятилитровых баллонах. 

Если пересечением отрезков $A$ и $B$ является пустое множество, то значения $N_1$ и $N_2$ противоречат друг другу и ответ на вопрос задачи равен $-1$.  
Иначе, $L = min(V_{A_l}, V_{B_l})$ - левая  граница пересечения отрезков $A$ и $B$, даст нам **минимальное** необходимое количество баллонов емкостью $60$ литров,а правая граница $R = max(V_{A_r}, V_{B_r})$ - **максимальное** необходимое количество шестидесятилитровых баллонов.  
Решением задачи в этом случае будут все целочисленные значения от $L$ до $R$ включительно.


### Код на языке программирования Python:

```python
n1 = int(input())
n2 = int(input())

a_l = n1 * 50 - 49
a_r = n1 * 50

b_l = n2 * 70 - 69
b_r = n2 * 70

l = max(a_l, b_l)
r = min(a_r, b_r)

if l > r:
    print(-1)
else:
    v1 = (l + 59) // 60
    v2 = (r + 59) // 60
    print(*[v for v in range(v1, v2 + 1)])
```
